```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(25)
```

```python
y = x ** 2
```

```python
plt.plot(x,y,'r--')
plt.xlabel('x Label')
plt.ylabel('y Label')
plt.title('My first plot')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')

plt.subplot(2,2,2)
plt.plot(y,x,'g')

plt.subplot(2,2,3)
plt.plot(x,y,'r')

plt.subplot(2,2,4)
plt.plot(y,x,'b--')
```
